```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import ShuffleSplit
from sklearn import linear_model
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
ss = ShuffleSplit(n_splits=1,
                 train_size=0.5,
                 test_size=0.5,
                 random_state=0)
```

```python
clf = linear_model.LogisticRegression(solver='liblinear')
```

```python
train_index, test_index = next(ss.split(X))

X_train, X_test = X[train_index], X[test_index]
y_train, y_test = y[train_index],y[test_index]
```

```python
clf.fit(X_train, y_train);
```

```python
print(clf.score(X_train, y_train))
```

```python
print(clf.score(X_test, y_test))
```

```python
ss = ShuffleSplit(n_splits=10,
                 train_size=0.5,
                 test_size=0.5,
                 random_state=0)

scores = []
for train_index, test_index in ss.split(X):
    
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    scores.append(score)
    
scores = np.array(scores)
print(scores)
```

```python
scores.mean()
```

```python
scores.std()
```

```python
print("{0} +/- {1}".format(scores.mean(), scores.std()))
```

```python
print("{0:4.2f} +/- {1:4.2f} %".format(scores.mean() * 100, scores.std() * 100))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_sizes = np.arange(0.1, 1.0, 0.1)
```

```python
all_mean = []
all_std = []

for train_size in train_sizes:
    
    ss = ShuffleSplit(n_splits=100,
                     train_size=train_size,
                     test_size=1-train_size)
    
    scores = []
    for train_index,test_index in ss.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        scores.append(score)
        
    scores = np.array(scores)
    print("train_size {0:.0f}%:{1:4.2f} +/- {2:4.2f} %".format(train_size *100,
                                                               scores.mean() *100,
                                                               scores.std() *100))
    all_mean.append(scores.mean() * 100)
    all_std.append(scores.std() * 100)
```

```python
plt.plot(train_sizes, all_mean)
plt.show()
```

```python
plt.plot(train_sizes, all_mean)
plt.ylim(70,100)
plt.xlim(0,1)
plt.show()
```

```python
plt.plot(train_sizes, all_mean)
plt.ylim(70, 100)
plt.xlim(0,1)
plt.errorbar(train_sizes, all_mean, yerr=all_std)
plt.xlabel("training size [%]")
plt.ylabel("recognition rate")
plt.title("Average of 10 hold-out tests for different training size")
plt.show()
```
