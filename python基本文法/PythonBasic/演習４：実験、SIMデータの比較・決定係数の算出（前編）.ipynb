```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("ggplot")
```

```python
df_sim = pd.read_csv("./data/SIM-Reaction-1.csv",delimiter=",",header=9, index_col=0)
df_sim2 = pd.read_csv("./data/SIM-Reaction-2.csv",delimiter=",",header=9, index_col=0)
```

```python
df_sim.head
```

```python
df_sim2.head()
```

```python
df_sim3 = pd.concat([df_sim, df_sim2])
```

```python
df_sim3
```

```python
df_ex = pd.DataFrame([])
for i in range(1,4):
    df = pd.read_excel("./data/exp-Reaction.xlsx", sheet_name="Sheet"+str(i),header=5,index_col=0)
    df["id"] = i
    df_ex = pd.concat([df_ex,df])
```

```python
df_ex
```

```python
groups = df_ex.groupby("id")
plt.figure(figsize=(12,9))
species = ["A","B","C","D"]
for name,group in groups:
    plt.subplot(2,2,name)
    for i in species:
        plt.scatter(group.index, group[i+"(exp)"],label="")
        plt.plot(df_sim3.index, df_sim3[i])
    plt.legend()
```

```python
group_t = df_ex.groupby(level="Time[min]")
```

```python
species_ex = ["A(exp)","B(exp)","C(exp)","D(exp)"]
df_ave = group_t[species_ex].mean()
```

```python
df_ave.head()
```

```python
df_errors = pd.DataFrame()
for i in species:
    diff = (df_ave[i+"(exp)"] - df_sim3[i]) ** 2
    df_errors[i] = diff
```

```python
df_errors
```

```python
df_errors = df_errors.dropna()
```

```python
df_errors
```

```python
df_num = df_errors.sum()
```

```python
denom = df_ave.var(ddof=False) * df_ave.shape[0]
```

```python
denom
```

```python
for i, j in enumerate(species):
    print(j, 1.0 - df_num[j] / denom[j+"(exp)"])
```
