```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_e = pd.read_excel("./data/data-for-graph.xlsx",sheet_name="exp", index_col=0)
df_s = pd.read_excel("./data/data-for-graph.xlsx",sheet_name="sim", index_col=0)
```

```python
df_e.head()
```

```python
plt.scatter(df_e.index, df_e["Number"])
plt.plot(df_s.index, df_s["Number"])
```

```python
plt.style.use("ggplot")
```

```python
plt.scatter(df_e.index, df_e["Number"])
plt.plot(df_s.index, df_s["Number"])
```

```python
plt.scatter(df_e.index, df_e["Number"], label="exp")
plt.plot(df_s.index, df_s["Number"], label="sim")
plt.xlabel("Time[Day]", fontsize = 18)
plt.ylabel("Number[-]", fontsize = 18)
plt.legend()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.scatter(df_e.index, df_e["Number"], label="exp")
ax1.plot(df_s.index, df_s["Number"], label="sim")
ax1.set_ylabel("Number[-]")
ax1.legend(bbox_to_anchor=(1.25,1.0))
ax2 = ax1.twinx()
ax2.set_ylim([0,35])
ax2.grid(False)
ax2.set_ylabel("Temp")
ax2.scatter(df_e.index, df_e["Temp"], color="blue", marker="^")
```
