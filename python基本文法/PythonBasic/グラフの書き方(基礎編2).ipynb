```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-10,10,0.01)
y = 1.0 / (1.0 + np.exp(-x))
```

```python
x2 = np.arange(-6,6,0.01)
y2 = np.sin(x2)
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x2,y2)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x2,y2)
```
